Now, run this work with all authorids. In df0, each column represents each professor. If there are more than one column, it means that this professor has multiple authorids with one name. Therefore, in this project, I would add up number of citations and consider them as their total number of citations.
* (There might be other people who have same names, but I find it almost impossible to tell which authorids belong to UIUC authors, so I would think of them as one person).

```python
total_citations = []

for _, row in df0.iterrows():
    num_citations = []
    for cell in row:
        if isinstance(cell, dict):
            author_id = cell.get('authorId')
            if author_id:
                url = f"https://api.semanticscholar.org/graph/v1/author/{author_id}/papers"
                response = requests.get(url, headers=headers)
                if response.status_code == 200:
                    data = response.json().get('data', [])
                    paper_ids = [paper["paperId"] for paper in data]
                    for paper_id in paper_ids:
                        url = f"https://api.semanticscholar.org/graph/v1/paper/{paper_id}?fields=url,year,authors,citations"
                        response = requests.get(url, headers=headers)
                        if response.status_code == 200:
                            citations = response.json().get('authors', [])
                            num_citations.append(len(citations))
                        else:
                            print(f"Failed to fetch paper details for paper ID: {paper_id}")
                else:
                    print(f"Failed to fetch papers for author ID: {author_id}")
    
    total_citations.append(sum(num_citations))

df_total_citations = pd.DataFrame({'total_citations': total_citations})     

df_total_citations
```

Add this result into the dataframe that we created before.

```python
profs.reset_index(drop=True, inplace=True)
df_total_citations.reset_index(drop=True, inplace=True)
profs["total_citations"] = df_total_citations["total_citations"]
profs
```
